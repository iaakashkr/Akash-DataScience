```python
# used dictionary to fill the Na value with 'Nothing' 
data_frame.fillna({'Effect: Impact to flight': 'Nothing'}, inplace=True) 
#used dictionary to fill the Na value with 'None'
data_frame.fillna({'Conditions: Precipitation': 'None'},inplace=True)   
# used dictionary to fill the Na value with 'No Remarks' 
data_frame.fillna({'Remarks': 'No Remarks'},inplace=True) 
# used dictionary to fill the Na value with 'Unknown'
data_frame.fillna({'Airport: Name':'Unknown'},inplace=True) 
# used dictionary to fill the Na value with 'Unknown'
data_frame.fillna({'Origin State':'Unknown'},inplace=True) 
# used dictionary to fill the Na value with 'Unknown'
data_frame.fillna({'Aircraft: Airline/Operator':'Unknown'},inplace=True) 
```

```python
 #used the 'for loop' to make an list to jump over the columns
 for column in ['Aircraft: Type', 'Altitude bin',
 'When: Phase of flight','Aircraft: Number of engines?',
 'Pilot warned of birds or wildlife?', 'Is Aircraft Large?', 'Feet above ground']:
    mode_value = data_frame[column].mode()[0]
 data_frame[column] = data_frame[column].fillna(mode_value)
```

```python
# using the if-else function to replace the missing cells in the 'Wildlife: Number struck'
def fill_missing_struck(row):
    # If the 'Wildlife: Number struck' column is not NaN, just return its value
    if pd.notna(row['Wildlife: Number struck']):
        return row['Wildlife: Number struck']
    # If 'Wildlife: Number struck' is NaN,  return based on 'Wildlife: Number Struck Actual'
    else:
        if row['Wildlife: Number Struck Actual'] > 100:
            return 'Over 100'
        elif 10 < row['Wildlife: Number Struck Actual'] <= 100:
            return '11 to 100'
        else:
            return '2 to 10'
data_frame['Wildlife: Number struck'] = data_frame.apply(fill_missing_struck, axis=1)
```

```python
#droping the rows with most of the  missing data.
data_frame.dropna(subset=['Aircraft: Number of engines?','FlightDate'], inplace=True)
```

<h1>Top 10 Birds Struck on the following Airport<h1>

```python
# Group data by airport name and sum up the number of bird strikes
airport_strikes = data_frame.groupby('Airport: Name')['Wildlife: Number Struck Actual'].sum()

# Plotting
plt.figure(figsize=(10, 6))
top_airports = airport_strikes.sort_values(ascending=False).head(10)
top_airports.plot(kind='bar', color='skyblue')

# Adding titles and labels
plt.title('Top 10 Airports by Bird Strikes')
plt.xlabel('Airport')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)

# Adding values on top of bars
for i, count in enumerate(top_airports):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

<h4>INSIGHTS:-<h4>
<h5>The highest no. of Birds strucks on the Airports was DALLAS/FORT WORTH INTL ARPT with an count of "2932", as the barplot show the other Airports with there birds struck's count.<h5>

```python
# Assuming we have a column 'Origin State' indicating the state of occurrence
bird_strikes_us = data_frame[data_frame['Origin State'].isin(['New York', 'Texas', 'Louisiana', 'Washington', 'Virginia',
       'Unknown', 'Delaware', 'DC', 'Georgia', 'Florida', 'California',
       'Illinois', 'Connecticut', 'Missouri', 'Rhode Island', 'Hawaii',
       'Arizona', 'Tennessee', 'South Carolina', 'South Dakota',
       'New Jersey', 'Colorado', 'Minnesota', 'Alabama', 'Ohio',
       'Wisconsin', 'Michigan', 'Massachusetts', 'Alaska',
       'North Carolina', 'Kentucky', 'Indiana', 'Oregon', 'Pennsylvania',
       'New Hampshire', 'Arkansas', 'Nevada', 'Mississippi', 'Maryland',
       'Maine', 'Quebec', 'Idaho', 'British Columbia', 'Utah', 'Nebraska',
       'Iowa', 'New Mexico', 'West Virginia', 'Oklahoma', 'North Dakota',
       'Vermont', 'Wyoming', 'Kansas', 'Prince Edward Island', 'Montana',
       'Puerto Rico', 'Ontario', 'Virgin Islands',
       'Newfoundland and Labrador', 'Alberta', 'Saskatchewan'])] # Include all relevant states

# Group by year and count the number of bird strikes
yearly_bird_strikes_us = bird_strikes_us.groupby(bird_strikes_us['FlightDate'].dt.year).size()

# Plot the yearly analysis of bird strikes in the US
plt.figure(figsize=(10, 6))
bars = yearly_bird_strikes_us.plot(kind='bar', color='skyblue')
plt.title('Yearly Analysis of Bird Strikes in the US')
plt.xlabel('Year')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding labels to each bar
for bar in bars.patches:
    bars.annotate(format(bar.get_height(), '.0f'), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), 
                   ha='center', 
                   va='center', 
                   xytext=(0, 5), 
                   textcoords='offset points')

plt.show()
```

```python
# Assuming we have a column 'Aircraft: Airline/Operator' indicating the airline/operator
# Group by airline/operator and count the number of bird strikes
top_airlines = data_frame['Aircraft: Airline/Operator'].value_counts().head(10)

# Plot the top 10 US airlines with the most bird strikes
plt.figure(figsize=(10, 6))  # Set figure size
bars = top_airlines.plot(kind='bar', color='skyblue')  # Plot bar chart
plt.title('Top 10 US Airlines by Bird Strikes')  # Set title
plt.xlabel('Airline/Operator')  # Set x-axis label
plt.ylabel('Number of Bird Strikes')  # Set y-axis label
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for y-axis

# Adding labels to each bar
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2,  # x-coordinate
             bar.get_height(),  # y-coordinate
             f'{int(bar.get_height())}',  # Text to display (convert to integer)
             ha='center',  # Horizontal alignment
             va='bottom')  # Vertical alignment

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
# Assuming we have a column 'FlightDate' indicating the flight date and a column 'Cost: Total $' indicating the cost
# Convert 'FlightDate' to datetime if it's not already in datetime format
data_frame['FlightDate'] = pd.to_datetime(data_frame['FlightDate'])

# Group by year and sum up the total cost for each year
yearly_cost = data_frame.groupby(data_frame['FlightDate'].dt.year)['Cost: Total $'].sum()

# Plot the yearly cost incurred due to bird strikes
plt.figure(figsize=(10, 6))
bars = yearly_cost.plot(kind='bar', color='skyblue')
plt.title('Yearly Cost Incurred due to Bird Strikes')
plt.xlabel('Year')
plt.ylabel('Cost ($)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding labels to each bar
for bar in bars.patches:
    plt.annotate(format(bar.get_height(), '.2f'), 
                 (bar.get_x() + bar.get_width() / 2, 
                  bar.get_height()), 
                 ha='center', 
                 va='center', 
                 xytext=(0, 5), 
                 textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
# Assuming we have a column 'Altitude' indicating the altitude of airplanes at the time of strike
plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(data_frame['Altitude bin'], bins=20, color='skyblue', edgecolor='black')
plt.title('Altitude of Airplanes at the Time of Bird Strike')
plt.xlabel('Altitude (feet)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding frequency labels above each bar
for frequency, bin_value, patch in zip(n, bins, patches):
    plt.text(bin_value + (bins[1] - bins[0]) / 2, frequency + 5, str(int(frequency)), ha='center', va='bottom')

plt.show()
```
